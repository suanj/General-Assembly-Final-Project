```python
# code for converting duration values more readable, duration(seconds).
import datetime


def get_isosplit(s, split):
    if split in s:
        n, s = s.split(split)
    else:
        n = 0
    return n, s


def parse_isoduration(s):
        
    # Remove prefix
    s = s.split('P')[-1]
    
    # Step through letter dividers
    days, s = get_isosplit(s, 'D')
    _, s = get_isosplit(s, 'T')
    hours, s = get_isosplit(s, 'H')
    minutes, s = get_isosplit(s, 'M')
    seconds, s = get_isosplit(s, 'S')

    # Convert all to seconds
    dt = datetime.timedelta(days=int(days), hours=int(hours), minutes=int(minutes), seconds=int(seconds))
    return int(dt.total_seconds())
```

```python
# insert data into the column.
for i in set(youtube.duration):
    x = parse_isoduration(i)
    youtube['duration'] = youtube['duration'].replace(i, x)
```

To have a numerical value for data analysis, 
I decided to convert values in "tags" into the number of tags in the video content.

```python
# this cell shows the type of each value in tags column and the value itself.
for i in set(youtube.tags):
    print(type(i))
    print(i)
```

```python
# code for converting tags values into the number of tags in a video content. 
count = 0

tags = set(youtube.tags)

for i in tags:
    ln = len(i)
    if i == '0':
        youtube['tags'] = youtube['tags'].replace(i, 0)
    elif ln > 1:
        for k in range(ln):
            if i[k] == ',':
                count += 1
            elif i[k] == ']':
                count += 1
                youtube['tags'] = youtube['tags'].replace(i, count)
    count = 0
```
