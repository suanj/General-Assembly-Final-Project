Import json files and scrape data. 

```python
import json
import pandas as pd 

file = "./Youtube_scraped_data/한번_들어봐.json"
data = None
with open(file, 'r') as f:
    data = json.load(f)

#print channe_id 
channel_id, stats = data.popitem()
print("channel_id:",channel_id)
channel_stats = stats["channel_statistics"]
video_stats = stats['video_data']
```

```python
#channel statistics
print('view', channel_stats["viewCount"])
print('subscriber', channel_stats.get("subscriberCount"))
print('videos', channel_stats["videoCount"])
```

```python
#videos satistics 
sorted_vids = sorted(video_stats.items(), key=lambda item: int(item[1]['viewCount']), reverse=True)
stats =[]
for vid in sorted_vids:
    video_id = vid[0]
    publish_time = vid[1]["publishedAt"]
    title = vid[1]["title"]
    channelId = vid[1]["channelId"]
    channelTitle = vid[1]["channelTitle"]
    tags = vid[1].get("tags")
    views = int(vid[1]["viewCount"])
    likes = int(vid[1]["likeCount"])
    dislikes = int(vid[1]["dislikeCount"])
    favorites = int(vid[1]["favoriteCount"])
    comments = vid[1].get('commentCount')
    duration = vid[1]["duration"]
    stats.append([video_id, publish_time, title, channelId, channelTitle, tags, views, likes, dislikes, favorites, comments, duration])
stats
```

```python
#create new data frame 
tfc = pd.DataFrame(stats, columns=["video_id", "publish_time", "title", "channel_Id", "channel_title", "tags", "views_count", "likes_count", "dislikes_count", "favorites_count", "comments_count", "duration"])
tfc
```

```python
#save dataset 
tfc.to_csv('한번 들어봐.csv')
```
